```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv("Test.csv")
train = pd.read_csv("Train.csv")
var_used = pd.read_csv("VariablesUsed.csv")
```

```python
train['MRP'] = pd.to_datetime(train['MRP'], format='%m/%d/%Y')
train["MRP_year"] = train["MRP"].apply(lambda x:x.year)
train["MRP_month"] = train["MRP"].apply(lambda x:x.month)
train["MRP_day"] = train["MRP"].apply(lambda x:x.day)
print(train.MRP_year.unique(), train.MRP_month.unique(), train.MRP_day.unique())
```

```python
train.drop(["MRP","MRP_year","MRP_day"],axis=1,inplace=True)
```

```python
train['ODATE'] = pd.to_datetime(train['ODATE'], format='%m/%d/%Y')
train["ODATE_year"] = train["ODATE"].apply(lambda x:x.year)
train["ODATE_month"] = train["ODATE"].apply(lambda x:x.month)
train["ODATE_day"] = train["ODATE"].apply(lambda x:x.day)
print(train.ODATE_year.unique(), train.ODATE_month.unique(), train.ODATE_day.unique())
```

```python
train.drop(["ODATE","ODATE_day"],axis=1,inplace=True)
```

```python
train['FPD'] = pd.to_datetime(train['FPD'], format='%m/%d/%Y')
train["FPD_year"] = train["FPD"].apply(lambda x:x.year)
train["FPD_month"] = train["FPD"].apply(lambda x:x.month)
train["FPD_day"] = train["FPD"].apply(lambda x:x.day)
print(train.FPD_year.unique(), train.FPD_month.unique(), train.FPD_day.unique())
```

```python
train.drop(["FPD","FPD_day"],axis=1,inplace=True)
```

```python
train.isna().sum()
```

```python
NANCol = []
for col in train:
    if train[col].isna().any():
        NANCol.append(col)
print(NANCol)
for name in NANCol:
    print(train[name].dtypes)
```

```python
DicFill = {
    'LOTV'  : train['LOTV'].median(skipna=True),
    'CLTV'  : train['CLTV'].median(skipna=True), 
    'NB'    : train['NB'].median(skipna=True), 
    'DTI'   : train['DTI'].median(skipna=True), 
    'BCSAO' : train['BCSAO'].median(skipna=True), 
    'CBCSAO': train['CBCSAO'].median(skipna=True), 
    'FTHBI' : train['FTHBI'].value_counts().index[0], 
    'MIP'   : train['MIP'].median(skipna=True), 
    'MIT'   : train['MIT'].median(skipna=True),
}
```

```python
train.fillna(DicFill, inplace = True)
print(train['FORECLOSED'].value_counts())
train
```

```python
from pandas.api.types import is_numeric_dtype
NumCol = []
NonNumcol = []
for col in train:
    if is_numeric_dtype(train[col]):
        NumCol.append(col)
    else:
        NonNumcol.append(col)
print(NumCol)
```

```python
train.reset_index(drop=True, inplace=True)
```

```python
# train["lNMONTHS"]=train.NMONTHS.apply(np.log)
N=train.shape[0]
perm=np.random.permutation(range(N))
train=train.loc[perm]
Ntrain=int(.8*N)
Ntest=N-Ntrain
train_train=train.iloc[0:Ntrain]
train_test=train.iloc[Ntrain:N]
```

```python
train_test = pd.concat([train_test, LP_U], axis=0)
train_test.reset_index(inplace=True)
```

```python
print(NumCol)
```

```python
import patsy as ps
from sklearn.linear_model import LinearRegression

formula="NMONTHS~CHANNEL+SELLER+FTHBI+LP+PTYPE+OCCSTAT+STATE+RMI+np.log(ORATE)+CRATE+np.log(OUPB)+LA+np.log(BCSAO)+CBCSAO+np.log(NUNITS)+np.log(MIP)+MIT+MRP_month+ODATE_year+ODATE_month+FPD_year"
Ytrain_train,Xtrain_train=ps.dmatrices(formula,train_train)
Ytrain_test,Xtrain_test=ps.dmatrices(formula,train_test)
Ntest=Ytrain_test.shape[0]
reg= LinearRegression().fit(Xtrain_train,Ytrain_train)
Ytest_pred=reg.predict(Xtrain_test)
e=pd.Series((Ytrain_test-Ytest_pred).reshape((Ntest,)))
MAD=e.apply(abs).mean()
print(MAD)
```
