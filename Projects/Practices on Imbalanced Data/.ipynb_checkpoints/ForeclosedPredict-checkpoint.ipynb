```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv("Test.csv")
train = pd.read_csv("Train.csv")
var_used = pd.read_csv("VariablesUsed.csv")
```

```python
train['MRP'] = pd.to_datetime(train['MRP'], format='%m/%d/%Y')
train["MRP_year"] = train["MRP"].apply(lambda x:x.year)
train["MRP_month"] = train["MRP"].apply(lambda x:x.month)
train["MRP_day"] = train["MRP"].apply(lambda x:x.day)
print(train.MRP_year.unique(), train.MRP_month.unique(), train.MRP_day.unique())
```

```python
train.drop(["MRP","MRP_year","MRP_day"],axis=1,inplace=True)
```

```python
train['ODATE'] = pd.to_datetime(train['ODATE'], format='%m/%d/%Y')
train["ODATE_year"] = train["ODATE"].apply(lambda x:x.year)
train["ODATE_month"] = train["ODATE"].apply(lambda x:x.month)
train["ODATE_day"] = train["ODATE"].apply(lambda x:x.day)
print(train.ODATE_year.unique(), train.ODATE_month.unique(), train.ODATE_day.unique())
```

```python
train.drop(["ODATE","ODATE_day"],axis=1,inplace=True)
```

```python
train['FPD'] = pd.to_datetime(train['FPD'], format='%m/%d/%Y')
train["FPD_year"] = train["FPD"].apply(lambda x:x.year)
train["FPD_month"] = train["FPD"].apply(lambda x:x.month)
train["FPD_day"] = train["FPD"].apply(lambda x:x.day)
print(train.FPD_year.unique(), train.FPD_month.unique(), train.FPD_day.unique())
```

```python
train.drop(["FPD","FPD_day"],axis=1,inplace=True)
```

```python
train.isna().sum()
```

```python
NANCol = []
for col in train:
    if train[col].isna().any():
        NANCol.append(col)
print(NANCol)
for name in NANCol:
    print(train[name].dtypes)
```

```python
DicFill = {
    'LOTV'  : train['LOTV'].median(skipna=True),
    'CLTV'  : train['CLTV'].median(skipna=True), 
    'NB'    : train['NB'].median(skipna=True), 
    'DTI'   : train['DTI'].median(skipna=True), 
    'BCSAO' : train['BCSAO'].median(skipna=True), 
    'CBCSAO': train['CBCSAO'].median(skipna=True), 
    'FTHBI' : train['FTHBI'].value_counts().index[0], 
    'MIP'   : train['MIP'].median(skipna=True), 
    'MIT'   : train['MIT'].median(skipna=True),
}
```

```python
train['FORECLOSED'] = train['FORECLOSED'].replace(True,1)
train['FORECLOSED'] = train['FORECLOSED'].replace(False,0)
```

```python
train.fillna(DicFill, inplace = True)
print(train['FORECLOSED'].value_counts())
train
```

```python
train.reset_index(drop=True, inplace=True)
```

```python
LP_U = train[train.LP == 'U']
LP_U
```

```python
from imblearn.over_sampling import RandomOverSampler

X = train.drop('FORECLOSED', axis=1)
y = train['FORECLOSED']
ros = RandomOverSampler(random_state=0)
X_resampled, y_resampled = ros.fit_resample(X, y)
X_resampled['FORECLOSED'] = y_resampled
train_new = X_resampled
```

```python
# train["lNMONTHS"]=train.NMONTHS.apply(np.log)
N=train_new.shape[0]
perm=np.random.permutation(range(N))
train_new=train_new.loc[perm]
Ntrain=int(.8*N)
Ntest=N-Ntrain
train_train=train_new.iloc[0:Ntrain]
train_test=train_new.iloc[Ntrain:N]
```

```python
train_test = pd.concat([train_test, LP_U], axis=0)
train_test.reset_index(inplace=True)
```

```python
from pandas.api.types import is_numeric_dtype
NumCol = []
NonNumcol = []
for col in train:
    if is_numeric_dtype(train[col]):
        NumCol.append(col)
    else:
        NonNumcol.append(col)
print(NumCol)
```

```python
import patsy as ps
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import RandomOverSampler

formula="FORECLOSED~CHANNEL+SELLER+FTHBI+LP+PTYPE+OCCSTAT+STATE+RMI+ORATE+CRATE+np.log(OUPB)+OLTERM+RMTM+LOTV+NB+DTI+BCSAO+CBCSAO+ODATE_month"
Ytrain_train,Xtrain_train=ps.dmatrices(formula,train_train)
Ytrain_test,Xtrain_test=ps.dmatrices(formula,train_test)
Logreg = LogisticRegression(solver = "newton-cg", max_iter=1000).fit(Xtrain_train,Ytrain_train.ravel())
Logreg.score(Xtrain_test,Ytrain_test)
```

```python
# Create new dataframe with prediction on prob
test['FORECLOSED'] = ''
Ytest,Xtest=ps.dmatrices(formula,test)
pred_foreclosed_test_prob = Logreg.predict_proba(Xtest)[:,1] # Prob of True
foreclosed_indexed_prob = pd.DataFrame(pred_foreclosed_test_prob, columns = ['Prob of True'])
# Sort and add column of Prediction
foreclosed_indexed_prob = foreclosed_indexed_prob.sort_values(by='Prob of True', ascending=False)
# Set the first 1k largest prob to True
def setTrue(i):
    if i < 1000:
        return True
    return False
foreclosed_indexed_prob.insert(1, 'FORECLOSED', [setTrue(i) for i in range(len(foreclosed_indexed_prob['Prob of True']))])
# Sort by index
foreclosed_indexed_prob = foreclosed_indexed_prob.sort_index(ascending=True)
# Drop prob
foreclosed_indexed_prob = foreclosed_indexed_prob.drop(columns=['Prob of True'])
# Combine LID
foreclosed_indexed_prob.insert(0, 'LID', test['LID'])

print(foreclosed_indexed_prob['FORECLOSED'].value_counts())
foreclosed_indexed_prob
```
